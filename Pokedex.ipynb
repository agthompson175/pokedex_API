```python
from IPython.display import clear_output as co
import requests
import pprint


class Pokemon:
    def __init__(self, name):
        self.name = name
        res = requests.get('https://pokeapi.co/api/v2/pokemon/{pokename}'.format(pokename=name))
        self.data = res.json()
        self.stats = {}
        
        
        self.stats['name'] = self.data['name']
        
        types=[]
        for poketype in self.data['types']:
            types.append(poketype['type']['name'])
        
        self.stats['types'] = types
        
        
        self.stats['height']=self.data['height']
        
        self.stats['weight']=self.data['weight']
        
        abilities = []
        for ability in self.data['abilities']:
            abilities.append(ability['ability']['name'])
        self.stats['ability'] = abilities
        
        statistics = {}
        for stattype in self.data['stats']:
            statistics[stattype['stat']['name']] = stattype['base_stat']
        self.stats['basestats'] = statistics
        
  #      self.stats['group_avgs'] = 
    

class Pokedex:
    @classmethod
    def run(self):
        print('Welcome to the PokeDex, you can add any pokemon to your Dex by name. Make sure you spell correctly! dummy...\n')
        done = False
        pokemon_total_list = []
        type_dict={}
        while not done:
            
            confirmation = input("Want to view a Pokemon? add/show/quit? ").lower()
            if confirmation == 'quit':
                done = True
            elif confirmation == 'add':
                co()
                try:
                    print('Welcome!')
                    pokemon_choice = input('Which Pokemon would you like to add to Pokedex? ').lower()

                    pokepoke=Pokemon(pokemon_choice)

                    pokemon_total_list.append(pokepoke)

                    print('\nYou have added the pokemon below to your PokeDex: \n\n' + str(pokemon_choice))
                except:
                    print('incorrect spelling')
            elif confirmation == 'show':
                co()
                print('your Pokedex is below:\n(grouped by type\n)')
                for pokemon in pokemon_total_list:
                    if pokemon.data['types'][0]['type']['name'] not in type_dict:
                        type_dict[pokemon.data['types'][0]['type']['name']] = [pokemon.stats]
                    else:
                        type_dict[pokemon.data['types'][0]['type']['name']].append(pokemon.stats)
                pprint.pprint(type_dict, sort_dicts = False)


                
                    
                                     
                    
Pokedex.run()




```
